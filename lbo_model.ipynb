Below is a step-by-step breakdown of the model.

### Step 2: Create Initial Dataset

The data includes revenue forecasts, EBITDA, free cash flows, interest payments, tax rates, debt outstanding, and equity contribution. The dataset also includes terminal value and multiples for exit calculations. I store this data in a Pandas DataFrame for easy manipulation. Column explanations are below.

```python
data = {
    'Company': ["Blake's Bakes Inc."] * 5,
    'Year': [1, 2, 3, 4, 5],
    'Revenue Forecast': [1000000, 1100000, 1200000, 1300000, 1400000],
    'EBITDA': [300000, 330000, 360000, 390000, 420000],
    'Free Cash Flow (FCF)': [200000, 220000, 240000, 260000, 280000],
    'Interest Payments': [50000, 48000, 46000, 44000, 42000],
    'Tax Rate': [0.25] * 5,
    'Discount Rate': [0.1] * 5,
    'Debt/Equity Ratio': [2] * 5,
    'Terminal Value': [5000000] + [None] * 4,
    'Debt Outstanding': [900000, 800000, 700000, 600000, 500000],
    'Equity Contribution': [300000] + [None] * 4,
    'Purchase Price': [1200000] + [None] * 4,
    'Multiples (Exit)': [8] + [None] * 4,
    'Exit Value': [None] * 4 + [3360000],
    'Equity Value at Exit': [None] * 4 + [2860000],
}
```

## Company

## Year
#### Represents the time horizon for the LBO model, from year 1 to year 5
#### Used as the basis for discounting future cash flows

## Revenue Forecast
#### The projected revenue for each year
#### Drives business growth and directly influences other metrics like EBITDA and free cash flow

## EBITDA (Earnings Before Interest, Taxes, Depreciation, and Amortization)
#### A key profitability metric showing the company's operating performance before non-operating expenses
#### Used to calculate free cash flow and determine the company’s valuation

## Free Cash Flow (FCF)
#### Represents the cash available after all operating expenses, taxes, and interest payments
#### Used to pay down debt or distribute to equity holders, making it critical for LBOs

## Interest Payments
#### Contains the interest payments on the company’s outstanding debt
#### Reduces the free cash flow available to equity holders

## Tax Rate
#### The effective tax rate applied to the company's earnings
#### Impacts the net profitability and post-tax cash flows

## Discount Rate
#### Used to calculate the present value of future cash flows
#### Accounts for the time value of money and investment risk, influencing enterprise value (EV)

## Debt/Equity Ratio
#### Represents the proportion of debt to equity used to finance the acquisition
#### Helps calculate how much of the purchase price is funded by debt versus equity

## Terminal Value
#### The estimated value of the company at the end of the projection period (year 5)
#### Assumes the company will continue operating indefinitely, used to calculate overall investment value

## Debt Outstanding
#### Tracks the amount of debt remaining to be paid off at the end of each year
#### Decreases over time as debt is repaid

## Equity Contribution
#### The amount of equity that the buyer contributes to the acquisition.
#### Represents the cash invested by equity holders, affecting their returns

## Purchase Price
#### The total purchase price for the company at the start of the model
#### Sum of debt and equity contributions, representing the total acquisition cost

## Multiples (Exit)
#### The exit multiple applied to EBITDA in the final year to calculate the exit value or sale price of the company
#### Reflects market conditions and perceived value of the company

## Exit Value
#### The sale price of the company at the end of the projection period, based on the exit multiple and EBITDA
#### Crucial for determining the overall success of the LBO

## Equity Value at Exit
#### The value of the equity portion at the time of the company’s sale
#### Calculated as the exit value minus remaining debt, representing the payoff to equity holders

## Step 3: Calculate Free Cash Flow (FCF)
### I calculate Free Cash Flow (FCF) by adjusting EBITDA for taxes, interest payments, and capital expenditures. FCF represents the cash available after operating expenses, which can be used to repay debt or distribute to equity holders.

```python
df['FCF'] = df['EBITDA'] * (1 - df['Tax Rate']) - df['Interest Payments']
```

```python
df
```

## Step 4: Present Value of Free Cash Flows (PV_FCF)
### I discount future free cash flows back to their present value using the discount rate provided. The sum of these discounted cash flows helps calculate the total enterprise value (EV).

```python
df['PV_FCF'] = df['FCF'] / (1 + df['Discount Rate']) ** df['Year']
```

```python
df
```

## Step 5: Discount Terminal Value
### The terminal value represents the company's value at the end of the LBO holding period. I discount this terminal value to present value using the discount rate.

```python
terminal_value_pv = df['Terminal Value'].iloc[0] / (1 + df['Discount Rate'].iloc[0]) ** df['Year'].iloc[-1]
```

## Step 6: Calculate Total Enterprise Value (EV)
### The total enterprise value is the sum of the present value of free cash flows and the present value of the terminal value. This gives us the value of the company today.

```python
total_ev = df['PV_FCF'].sum() + terminal_value_pv
```

## Step 7: Calculate Debt and Equity Contributions
### I calculate the debt and equity contributions based on the debt/equity ratio. This will show how much of the purchase price is financed through debt and equity.

```python
purchase_price = total_ev
df['Debt Contribution'] = purchase_price * df['Debt/Equity Ratio'].iloc[0] / (1 + df['Debt/Equity Ratio'].iloc[0])
df['Equity Contribution'] = purchase_price - df['Debt Contribution']
```

```python
df
```

```python
df.columns
```

## Step 8: Add Operating Assumptions
### I add detailed operating assumptions such as capital expenditures (CapEx), depreciation, and changes in working capital to reflect real-world financial complexities. These factors impact free cash flow and are essential for detailed financial modeling.

```python
df['CapEx'] = [-100000, -110000, -120000, -130000, -140000]
df['Depreciation'] = [50000, 60000, 70000, 80000, 90000]
df['Change in Working Capital'] = [-20000, -22000, -24000, -26000, -28000]
df['FCF'] = df['EBITDA'] * (1 - df['Tax Rate']) - df['Interest Payments'] - df['CapEx'] - df['Change in Working Capital']
```

```python
df
```

## Step 9: Create Debt Tranches
### I model different types of debt, including senior and subordinated debt, with respective interest rates and repayment schedules. This structure mimics the layered debt financing typical in leveraged buyouts.

```python
df['Senior Debt'] = [500_000, 450000, 400000, 350000, 300000]
df['Subordinated Debt'] = [400000, 350000, 300_000, 250_000, 200_000]
df['Total Debt Outstanding'] = df['Senior Debt'] + df['Subordinated Debt']
df['Interest Payments'] = df['Senior Debt'] * 0.05 + df['Subordinated Debt'] * 0.08
```

```python
df
```

## Step 10: Exit Value and Final Equity Value
### I calculate the exit value based on the EBITDA multiple at the end of the holding period. The final equity value is determined by subtracting any remaining debt from the exit value.

```python
exit_value = df['EBITDA'].iloc[-1] * df['Multiples (Exit)'].iloc[0]
debt_at_exit = df['Total Debt Outstanding'].iloc[-1]
equity_value_at_exit = exit_value - debt_at_exit
```

## Step 11: Add Equity Waterfall
### I divide the equity into preferred and common portions, with preferred equity receiving a 10% return before common equity holders receive their distributions.

```python
df['Preferred Equity Contribution'] = df['Equity Contribution'] * 0.7
df['Common Equity Contribution'] = df['Equity Contribution'] * 0.3
df['Preferred Equity Returns'] = df['Preferred Equity Contribution'] * 1.1  # 10% return to preferred equity
df['Common Equity Returns'] = equity_value_at_exit - df['Preferred Equity Returns']
```

```python
df
```

## Step 12: Calculate IRR and NPV
### I calculate the internal rate of return (IRR) and net present value (NPV) for the equity holders based on cash flows over the holding period. These metrics indicate the profitability of the investment.

```python
cash_flows = [-df['Equity Contribution'].iloc[0]] + df['FCF'].tolist() + [equity_value_at_exit]
irr = npf.irr(cash_flows)
npv = npf.npv(df['Discount Rate'].iloc[0], cash_flows)
```

## Step 13: Implement Dividend Recaps
### I simulate dividend recaps after year 3, where some of the free cash flow is distributed to equity holders rather than used for debt repayment.

```python
df['Dividend Payout'] = [0, 0, 50_000, 100_000, 150000]  # Example of dividends paid to equity holders
df['FCF_after_dividends'] = df['FCF'] - df['Dividend Payout']
```

## Step 14: Monte Carlo Simulation
### I run a Monte Carlo simulation to test different exit multiples (ranging from 6x to 10x) and calculate the resulting IRRs across 1,000 different scenarios. This simulation helps us assess the risk and variability of returns.

```python
n_simulations = 1_000
irr_results = []
exit_multiples = np.random.uniform(6, 10, n_simulations)

for multiple in exit_multiples:
    exit_value_sim = df['EBITDA'].iloc[-1] * multiple
    equity_value_at_exit_sim = exit_value_sim - debt_at_exit
    irr_sim = npf.irr([-df['Equity Contribution'].iloc[0]] + df['FCF'].tolist() + [equity_value_at_exit_sim])
    irr_results.append(irr_sim)

plt.hist(irr_results, bins=50)
plt.title('Distribution of IRRs from Monte Carlo Simulations')
plt.show()
```

## Sensitivity Analysis for Revenue Growth and CapEx

In this section, we conduct a **sensitivity analysis** to examine how variations in revenue growth rates and capital expenditures (CapEx) affect the Internal Rate of Return (IRR). Sensitivity analysis helps to understand the robustness of the investment by modeling different operational scenarios.

- **Growth Sensitivities:** We vary the revenue growth rate between 2% and 10%. Higher growth rates could reflect successful expansion efforts, while lower growth rates might indicate market challenges.
- **CapEx Sensitivities:** We vary CapEx growth rates between -2% and -10%, simulating different levels of investment in capital projects. Higher CapEx may indicate higher reinvestment in assets, while lower CapEx could reflect cost-cutting or more efficient capital allocation.

For each combination of growth rate and CapEx rate:
1. **Revenue Forecast:** We calculate the revenue forecast over the investment horizon, adjusting for the specified growth rate.
2. **CapEx:** We adjust CapEx over time based on the given rate.
3. **Free Cash Flow (FCF):** We recalculate the Free Cash Flow using the updated revenue and CapEx figures, while factoring in EBITDA, taxes, interest payments, and changes in working capital.
4. **IRR Calculation:** The IRR is calculated based on the new Free Cash Flow and the final equity value at exit.

The results of this sensitivity analysis show how the IRR varies with different combinations of revenue growth and CapEx, providing insights into which factors have the most significant impact on returns.

The printed output lists the IRR for each growth rate and CapEx combination, helping to identify favorable or risky operating conditions.

```python
growth_sensitivities = np.linspace(0.02, 0.10, 5)
capex_sensitivities = np.linspace(-0.02, -0.10, 5)

sensitivity_results = []

for growth_rate in growth_sensitivities:
    for capex_rate in capex_sensitivities:
        df['Revenue Forecast'] = [1000000 * (1 + growth_rate) ** year for year in df['Year']]
        df['CapEx'] = [-100000 * (1 + capex_rate) ** year for year in df['Year']]
        df['FCF'] = df['EBITDA'] * (1 - df['Tax Rate']) - df['Interest Payments'] - df['CapEx'] - df['Change in Working Capital']
        irr_sensitivity = npf.irr([-df['Equity Contribution'].iloc[0]] + df['FCF'].tolist() + [equity_value_at_exit])
        sensitivity_results.append((growth_rate, capex_rate, irr_sensitivity))

print("Sensitivity Analysis Results (Growth Rate, CapEx Rate, IRR):")
print()

for growth_rate, capex_rate, irr_sensitivity in sensitivity_results:
    print(f"Growth: {growth_rate:.2%}, CapEx: {capex_rate:.2%}, IRR: {irr_sensitivity:.2%}")
```

## Multiple Exit Strategy Analysis (Strategic Sale, IPO, Secondary Buyout)

In this section, we perform a **scenario analysis** to evaluate how different exit strategies affect the final equity value and Internal Rate of Return (IRR). The three exit scenarios considered are:

1. **Strategic Sale:** The company is sold to a strategic buyer, who might pay a premium for synergies or market share. We assume an exit multiple of 10x EBITDA.
2. **IPO (Initial Public Offering):** The company goes public, usually commanding a higher valuation due to access to a broader investor base. We assume an exit multiple of 12x EBITDA.
3. **Secondary Buyout:** Another private equity firm purchases the company, typically resulting in a lower exit multiple than an IPO but higher than a strategic sale. We assume an exit multiple of 8x EBITDA.

For each scenario:
- We calculate the **exit value** by multiplying the final year’s EBITDA by the exit multiple specific to that scenario.
- We then determine the **equity value at exit** by subtracting the outstanding debt from the exit value.
- Finally, we compute the **IRR** for each scenario to assess the potential returns under different exit strategies.

This scenario analysis helps illustrate the sensitivity of the investment’s return to various exit outcomes, giving insights into the potential upside and downside of different strategies.

```python
exit_scenarios = ['Strategic Sale', 'IPO', 'Secondary Buyout']
exit_multiples_by_scenario = {'Strategic Sale': 10, 'IPO': 12, 'Secondary Buyout': 8}

for scenario in exit_scenarios:
    exit_value_scenario = df['EBITDA'].iloc[-1] * exit_multiples_by_scenario[scenario]
    equity_value_scenario = exit_value_scenario - debt_at_exit
    irr_scenario = npf.irr([-df['Equity Contribution'].iloc[0]] + df['FCF'].tolist() + [equity_value_scenario])
    print(f"{scenario} - Exit Value: {exit_value_scenario}, IRR: {irr_scenario:.2%}")
```

## Stress Testing (Downside Protection)

In this section, we conduct a **worst-case scenario analysis** to stress test the LBO model. This analysis simulates a situation where the company's EBITDA declines by 20%, which could happen due to adverse market conditions, increased competition, or operational challenges.

- **Worst-Case EBITDA:** We begin by reducing the company's current EBITDA by 20% to reflect the impact of this downturn.
- **Worst-Case Free Cash Flow (FCF):** We then recalculate the Free Cash Flow (FCF) based on this reduced EBITDA, factoring in the same tax rate, interest payments, capital expenditures (CapEx), and changes in working capital.
- **Worst-Case Exit Value:** The exit value is recalculated using the worst-case EBITDA and applying the same exit multiple as in the base case scenario.
- **Worst-Case Equity Value:** From the recalculated exit value, we subtract the remaining debt at exit to determine the worst-case equity value at exit.
- **Worst-Case IRR:** Finally, we calculate the Internal Rate of Return (IRR) under these worst-case conditions to understand the potential downside risk.

This worst-case scenario analysis provides insights into how the investment could perform under unfavorable conditions, helping to gauge the resilience of the LBO model. The resulting worst-case IRR shows the return on investment in a stressed environment.

```python
worst_case_ebitda = df['EBITDA'] * 0.8  # 20% decline in EBITDA
worst_case_fcf = worst_case_ebitda * (1 - df['Tax Rate']) - df['Interest Payments'] - df['CapEx'] - df['Change in Working Capital']
worst_case_exit_value = worst_case_ebitda.iloc[-1] * df['Multiples (Exit)'].iloc[0]
worst_case_equity_value = worst_case_exit_value - debt_at_exit
worst_case_irr = npf.irr([-df['Equity Contribution'].iloc[0]] + worst_case_fcf.tolist() + [worst_case_equity_value])

print(f"Worst-Case Scenario IRR: {worst_case_irr:.2%}")
```

```python
# 1. Dynamic Debt/Equity Ratio and Interest Rate Sensitivity

# Dynamic Debt/Equity ratio based on changing leverage conditions
df['Debt/Equity Ratio'] = [2, 1.8, 1.6, 1.4, 1.2]  # Example of changing leverage over time

# Interest Rate Sensitivity (rising, stable, falling)
interest_rate_scenarios = {
    'Rising Rates': [0.05, 0.06, 0.07, 0.08, 0.09],
    'Stable Rates': [0.05] * 5,
    'Falling Rates': [0.09, 0.08, 0.07, 0.06, 0.05]
}

scenario_irrs_interest = {}
for scenario, rates in interest_rate_scenarios.items():
    df['Interest Payments'] = df['Senior Debt'] * rates + df['Subordinated Debt'] * [r * 1.5 for r in rates]
    df['FCF_after_dividends'] = df['EBITDA'] * (1 - df['Tax Rate']) - df['Interest Payments'] - df['CapEx'] - df['Change in Working Capital']
    scenario_irrs_interest[scenario] = npf.irr([-df['Equity Contribution'].iloc[0]] + df['FCF_after_dividends'].tolist() + [equity_value_at_exit])

# Display IRR results for each interest rate scenario
print("IRR Results for Interest Rate Scenarios:")
for scenario, irr in scenario_irrs_interest.items():
    print(f"{scenario}: IRR = {irr:.2%}")
```

```python
# 2. Recession Scenario with Revenue and EBITDA Compression

df['Recession Revenue Forecast'] = df['Revenue Forecast'] * 0.85  # 15% revenue drop in recession
df['Recession EBITDA'] = df['EBITDA'] * 0.80  # 20% EBITDA compression

recession_exit_value = df['Recession EBITDA'].iloc[-1] * df['Multiples (Exit)'].iloc[0] * 0.8  # 20% multiple compression
recession_equity_value = recession_exit_value - debt_at_exit
recession_irr = npf.irr([-df['Equity Contribution'].iloc[0]] + df['FCF_after_dividends'].tolist() + [recession_equity_value])

print(f"Recession Scenario IRR: {recession_irr:.2%}")
```

```python
# 3. Calculate Total and Equity IRRs

total_irr = npf.irr([-purchase_price] + df['FCF_after_dividends'].tolist() + [exit_value])
equity_irr = npf.irr([-df['Equity Contribution'].iloc[0]] + df['FCF_after_dividends'].tolist() + [equity_value_at_exit])

print(f"Total IRR: {total_irr:.2%}")
print(f"Equity IRR: {equity_irr:.2%}")
```

```python
# 4. Payback Period for Equity Investment

cumulative_cash_flow = np.cumsum([-df['Equity Contribution'].iloc[0]] + df['FCF_after_dividends'].tolist())
payback_year = np.argmax(cumulative_cash_flow >= 0)
print(f"Equity Payback Period: {payback_year} years")
```

```python
# 5. Waterfall Chart for Value Creation

value_creation = {
    'EBITDA Growth': df['Improved EBITDA Margin'].iloc[-1] - df['EBITDA'].iloc[-1],
    'Multiple Expansion': (exit_value - total_ev),
    'Debt Repayment': df['Total Debt Outstanding'].iloc[0] - df['Total Debt Outstanding'].iloc[-1],
    'Other Improvements': 0  # Placeholder for additional improvements
}

waterfall_fig = go.Figure(go.Waterfall(
    name="Value Creation Drivers",
    orientation="v",
    measure=["relative", "relative", "relative", "total"],
    x=["EBITDA Growth", "Multiple Expansion", "Debt Repayment", "Total Value Created"],
    y=[value_creation['EBITDA Growth'], value_creation['Multiple Expansion'], value_creation['Debt Repayment'], equity_value_at_exit]
))
waterfall_fig.update_layout(title="Waterfall Chart for Value Creation")
waterfall_fig.show()
```

```python
# 6. Real Option Valuation

def black_scholes_call(S, K, T, r, sigma):
    d1 = (np.log(S/K) + (r + 0.5 * sigma**2) * T) / (sigma * np.sqrt(T))
    d2 = d1 - sigma * np.sqrt(T)
    return S * norm.cdf(d1) - K * np.exp(-r * T) * norm.cdf(d2)

option_value = black_scholes_call(exit_value, purchase_price, 1, df['Discount Rate'].iloc[0], 0.2)
print(f"Real Option Value: ${option_value:,.2f}")
```

```python
# 7. Convert Annual Cash Flows to Quarterly Cash Flows

df['Quarterly FCF'] = df['FCF_after_dividends'] / 4  # Example of quarterly conversion
quarterly_cash_flows = df['Quarterly FCF'].tolist()
quarterly_cash_flows.insert(0, -df['Equity Contribution'].iloc[0])

quarterly_irr = npf.irr(quarterly_cash_flows)
print(f"Quarterly IRR: {quarterly_irr:.2%}")
```

```python
# 8. Covenant Breach Simulation

df['Leverage Covenant Breach'] = df['Leverage Ratio (Debt/EBITDA)'] > 4.0  # Example threshold
df['EBITDA Covenant Breach'] = df['EBITDA'] < 250000  # Example threshold

covenant_breach_years = df[df['Leverage Covenant Breach'] | df['EBITDA Covenant Breach']]['Year'].tolist()
print(f"Covenant Breach in Years: {covenant_breach_years}")
```

```python
# 9. Industry-Specific Metrics (Example: Subscription Business)

df['Churn Rate'] = [0.05, 0.06, 0.07, 0.08, 0.09]  # Example churn rate for subscription model
df['Subscriber Growth Rate'] = [0.10, 0.08, 0.06, 0.05, 0.04]  # Subscriber growth rate

# Calculate Subscribers and Revenue Based on Churn and Growth
df['Subscribers'] = [10000] + [(10000 * (1 + df['Subscriber Growth Rate'].iloc[i]) * (1 - df['Churn Rate'].iloc[i])) for i in range(1, len(df))]
df['Revenue from Subscribers'] = df['Subscribers'] * 10  # Assuming $10 per subscriber

print(df[['Year', 'Subscribers', 'Revenue from Subscribers']])
```
